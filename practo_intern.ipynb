```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Load the dataset
df =pd.read_csv(r"C:\Users\Peps\Documents\Downloads\heart_disease_health_indicators.csv")


# Display the first few rows of the dataset and the datatypes of columns
df.head(), df.info()
```

```python
# Convert relevant columns to categorical data type for optimization and analysis
categorical_columns = [
    'HeartDiseaseorAttack', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 'Diabetes',
    'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost',
    'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age', 'Education', 'Income'
]

# Convert categorical columns to 'category' type
df[categorical_columns] = df[categorical_columns].astype('category')

# Final check to ensure dataset is clean and ready for analysis
df.info(), df.head()
```

```python
import seaborn as sns
```

```python
# Set a common style for the plots
sns.set(style="whitegrid")
```

```python
# a. Plot histograms or bar charts for key variables (Age, BMI, MentHlth, PhysHlth)

# Plot Age Distribution (as categorical data, we'll use a count plot)
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=df)
plt.title('Age Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot BMI Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], bins=30, kde=True)
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Mental Health Days Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['MentHlth'].astype(int), bins=30, kde=True)
plt.title('Mental Health Days Distribution(in past 30 days)')
plt.xlabel('Number of MentHlth Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Physical Health Days Distribution
plt.figure(figsize=(10,6))
sns.histplot(df["PhysHlth"], bins=30, kde=True)
plt.title("Physical Health Days Distribution(in past 30 days)")
plt.xlabel('Number of PhysHlth Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Investigate Prevalence of Health Conditions (HighBP, HighChol)
# Bar plot for prevalence of high blood pressure and high cholesterol
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="HighBP", data=df)
plt.title('Prevalence of High Blood Pressure')
plt.xlabel("High Blood Pressure (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='HighChol', data=df)
plt.title('Prevalence of High Cholesterol')
plt.xlabel('High Cholesterol (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# c. Analyze Distribution of Heart Disease (Target Variable)
plt.figure(figsize=(10, 6))
sns.countplot(x='HeartDiseaseorAttack', data=df)
plt.title('Distribution of Heart Disease Cases')
plt.xlabel('Heart Disease or Attack (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# BIVARIATE ANALYSIS
```

```python
# Compare HighBP distribution between individuals with and without heart disease
plt.figure(figsize=(10, 6))
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df)
plt.title('High Blood Pressure and Heart Disease', pad=20, fontsize=16)
plt.xlabel('High Blood Pressure (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Check for null or non-integer values
print(df['HighBP'].unique())
print(df['HeartDiseaseorAttack'].unique())
```

```python
# Compare HighChol distribution between individuals with and without heart disease

plt.figure(figsize=(10, 6))
sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df)
plt.title('High Cholesterol and Heart Disease')
plt.xlabel('High Cholesterol (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Compare BMI distribution for individuals with and without heart disease
plt.figure(figsize=(10, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df)
plt.title('BMI Distribution and Heart Disease')
plt.xlabel('Heart Disease or Attack (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.show()
```

```python
# Scatter plot between BMI and Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='BMI', hue='HeartDiseaseorAttack', data=df)
plt.title('BMI vs Age (Colored by Heart Disease)')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

```python
# Scatter plot between MentHlth and PhysHlth
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MentHlth', y='PhysHlth', hue='HeartDiseaseorAttack', data=df)
plt.legend(loc='upper right')
plt.title('Mental Health Days vs Physical Health Days')
plt.xlabel('Number of Mental Health Days')
plt.ylabel('Number of Physical Health Days')
plt.show()
```

```python
# Compare heart disease prevalence by Age group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', hue='HeartDiseaseorAttack', data=df)
plt.title('Heart Disease by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Compare heart disease prevalence by Gender (Sex)
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='HeartDiseaseorAttack', data=df)
plt.title('Heart Disease by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```
